```python
#Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the dataset from the seaborn built-in datasets
import seaborn as sns
df = sns.load_dataset('car_crashes')
df.head()
```

```python
#We have different types of plots in seaborn
#1. Distribution Plots
#2. Categorical Plots
#3. Relational Plots
```

```python
#1. Distribution Plots
#Distribution plots are used to visualize the distribution of a single variable. 
#They help us understand the underlying distribution of the data, such as whether it is normal, skewed, or has multiple peaks.
sns.distplot
```

```python
sns.displot(df['not_distracted'])
plt.show()
```

```python
#Joint Plots
#Joint plots are used to visualize the relationship between two variables. 
#They combine a scatter plot with histograms or density plots to show the distribution of each variable along with their joint distribution.
sns.jointplot (x='alcohol', y='speeding', data=df, kind='hex')
plt.show()
```

```python
sns.jointplot (x='alcohol', y='speeding', data=df, kind='reg')
plt.show()
```

```python
sns.jointplot (x='alcohol', y='speeding', data=df, kind='kde')
plt.show()
```

```python
#2. KDE Plot
sns.kdeplot(df['alcohol'])
```

```python
#3.Pair Plot
sns.pairplot(df)
plt.show()
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
#Use pair plot to visualize the pairwise relationships in the tips dataset
sns.pairplot(tips_df, hue='sex')
plt.show()
```

```python
#Rug Plot
sns.rugplot(tips_df['tip'])
plt.show()
```

```python
#Styling in seaborn
#Seaborn provides several built-in themes to style your plots. You can use the set_style() function to apply a specific theme.
sns.set_style('whitegrid')
plt.figure(figsize =(12,9))
sns.displot(df['not_distracted'])
plt.show()
```

```python
#Categorical Plots
#Categorical plots are used to visualize the relationship between a categorical variable and a numerical variable. 
#They help us understand how the numerical variable varies across different categories of the categorical variable.

#1.Bar Plot
sns.barplot(x='sex', y='total_bill', data=tips_df)
plt.show()
```

```python
#2. Count Plot
sns.countplot(x='sex', data=tips_df)
plt.show()
```

```python
#3.Box Plot
#Box plots are used to visualize the distribution of a numerical variable across different categories of a categorical variable.
sns.boxplot(x='day', y='total_bill', data=tips_df ,hue='sex')
plt.show()
```

```python
#4.Violin Plot
#Violin plots are similar to box plots but also show the kernel density estimation of the numerical variable.
sns.violinplot(x='day', y='total_bill', data=tips_df ,hue='sex')
plt.show()
```

```python
#5. Strip Plot
#Strip plots are used to visualize the distribution of a numerical variable across different categories of a categorical variable.
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True,hue='sex',dodge=True)
plt.show()
```

```python
#Matrix Plot
#Matrix plots are used to visualize the relationship between multiple variables in a matrix format.
flights = sns.load_dataset('flights')
flights= flights.pivot_table(index='month', columns='year',values='passengers')
sns.heatmap(flights, cmap='Blues')
plt.show()
```
